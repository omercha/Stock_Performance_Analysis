## Introduction
This project analyses the stock performance of four well-known companies: **Tesla (TSLA)**, **Amazon (AMZN)**, **Advanced Micro Devices (AMD)** and **GameStop (GME)**. The goal is to fetch historical financial data for these companies, visualise their performance over time and generate a comprehensive dashboard that helps identify trends and patterns in their share prices. By examining these stocks, we aim to understand their behaviour in the market, compare their performance and observe the influence of key events on their share price over the past year.

## Data Collection

In this section, we will extract historical share price data for Tesla, Amazon, AMD and GameStop. We will use the **yfinance** library to obtain stock price information for each company over the last year. This data will be critical in visualising their performance and comparing trends.

```python
# Importing libraries
import yfinance as yf
import pandas as pd

# Defining the list of stocks that we are interested in
stocks = ['TSLA', 'AMZN', 'AMD', 'GME']
stock_data = {}

# Looping through each stock to fetch its data
for stock in stocks:
    ticker = yf.Ticker(stock)
    stock_data[stock] = ticker.history(period='1y')
```

## Data Visualisation

We will use **Seaborn** and **Matplotlib** to visualise the stock price trends over the past year for each company, specifically the **closing prices** and **trading volume**. This will help us observe the similarities and differences in the companies' stock market performance and provide insights into key events that may have influenced their stock prices.

```python
# Importing libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Closing prices
# Configuring plot
plt.figure(figsize=(14,8))
sns.set_theme(style='darkgrid')

# Looping through each stock to plot the closing prices
for stock in stocks:
     sns.lineplot(data=stock_data[stock], x=stock_data[stock].index, y='Close', label=stock)
  
# Adding titles and labels
plt.title('Stock Price Performance of Tesla, Amazon, AMD, and GameStop (Last Year)')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')

# Adding a key and a background grid
plt.legend()
plt.grid(True)

# Displays the closing prices
plt.show()


# Trading volumes
# Configuring plots
fig, axes = plt.subplots(nrows=len(stocks), ncols=1, figsize=(14, 8), sharex=True)

# Looping through each stock to plot the trading volumes separately
for i, stock in enumerate(stocks):
    axes[i].plot(stock_data[stock].index, stock_data[stock]['Volume'], label=stock, color='blue')
    axes[i].set_title(f'{stock} Trading Volume (Last Year)')
    axes[i].set_ylabel('Volume')
    axes[i].legend()
    axes[i].grid(True)

# Setting common x-axis label for all subplots
plt.xlabel('Date')

# Adjusting spacing
plt.tight_layout()

# Displays trading volumes
plt.show()
```

## Dashboard Summary

Based on the visualisations, we can draw some initial observations and insights:

1. **Tesla (TSLA)** and **Amazon (AMZN)** have shown steady growth, with occasional significant price spikes that correspond to important corporate events, earnings reports, or market trends.
